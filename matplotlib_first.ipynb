```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 4 * np.pi, 100)      # Create an array of 100 values evenly spaced between 0 and 2π.
y = np.sin(x)                           # Compute the sine of each value in x.

plt.figure(figsize=(10, 3.2))           # Create a new figure with a width of 8 inches and a height of 4 inches.
plt.plot(x, y, label='sin(x)', color='purple', linewidth=2.5)  # Plot y vs x with a blue line and a label.
plt.title('Simple Line Plot')           # Set the title of the plot.
plt.xlabel('x')                         # Label the x-axis.
plt.ylabel('sin(x)')                    # Label the y-axis.
plt.legend()                            # Display the legend on the plot.
plt.grid(True)                          # Enable the grid for better readability.
plt.show()                              # Render and display the line plot.
```

```python
np.random.seed(0)                        # Set seed for reproducibility; random numbers will be the same every time.
x_scatter = np.random.rand(50)           # Generate 50 random numbers for the x-axis.
y_scatter = np.random.rand(50)           # Generate 50 random numbers for the y-axis.
colors = np.random.rand(50)              # Generate 50 random numbers for the color mapping of each point.
sizes = 1000 * np.random.rand(50)        # Generate 50 random sizes (scaled up for visibility) for each point.
plt.figure(figsize=(8, 6))               # Create a new figure with a width of 8 inches and a height of 6 inches.
plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.7, cmap='viridis')  # Create a scatter plot with varying colors and sizes.
plt.colorbar()                           # Add a color bar to indicate the mapping of color values.
plt.title('Scatter Plot Example')        # Set the title of the scatter plot.
plt.xlabel('Random X')                   # Label the x-axis.
plt.ylabel('Random Y')                   # Label the y-axis.
plt.show()                               # Render and display the scatter plot.
```

```python
data = np.random.randn(1000)            # Generate 1000 data points from a standard normal distribution.
plt.figure(figsize=(10, 5))             # Create a new figure with a width of 8 inches and a height of 6 inches.
plt.hist(data, bins=100, color='green', alpha=1)  # Plot a histogram with 30 bins, green bars, and set transparency.
plt.title('Histogram Example')          # Set the title of the histogram.
plt.xlabel('Value')                     # Label the x-axis.
plt.ylabel('Frequency')                 # Label the y-axis.
plt.show()                              # Render and display the histogram.
```

```python
x = np.linspace(0, 10, 100)             # Create an array of 100 values evenly spaced between 0 and 10.
y1 = np.sin(x)                          # Compute the sine of each value in x for the first subplot.
y2 = np.cos(x)                          # Compute the cosine of each value in x for the second subplot.
fig, axs = plt.subplots(1, 2, figsize=(11, 3.5))  # Create a figure with a 1x2 grid of subplots and set the figure size.
axs[0].plot(x, y1, 'tab:blue')          # Plot the sine function on the first subplot with a tab blue color.
axs[0].set_title('Sine')                # Set the title for the first subplot.
axs[0].set_xlabel('x')                  # Label the x-axis on the first subplot.
axs[0].set_ylabel('sin(x)')             # Label the y-axis on the first subplot.
axs[0].grid(True)
axs[1].plot(x, y2, 'tab:orange')        # Plot the cosine function on the second subplot with a tab orange color.
axs[1].set_title('Cosine')              # Set the title for the second subplot.
axs[1].set_xlabel('x')                  # Label the x-axis on the second subplot.
axs[1].set_ylabel('cos(x)')             # Label the y-axis on the second subplot.
axs[1].grid(True)
plt.tight_layout()                      # Adjust the layout to prevent overlap between subplots.
plt.show()                              # Render and display the subplots.
```

```python
import numpy as np                               # Import NumPy for numerical operations.
import matplotlib.pyplot as plt                  # Import Matplotlib for plotting.
from sklearn.linear_model import LinearRegression# Import the Linear Regression model.

X      = np.array([[1], [2], [3], [4], [5]])     # Feature data: 5 samples, one feature each.
y      = np.array([1, 3, 2, 3, 5])               # Target data: corresponding values for each sample.
model  = LinearRegression()                      # Initialize the Linear Regression model.
model.fit(X, y)                                  # Train the model using the provided data.
y_pred = model.predict(X)                        # Predict target values based on input features.

plt.figure(figsize=(8, 5))                       # Create a figure with width 8 and height 6 inches.
plt.scatter(X, y, color='blue', label='Data')    # Plot the original data points in blue.
plt.plot(X, y_pred, color='red', linewidth=2, label='Predicted Line')  # Plot the regression line in red.
plt.title('Simple Linear Regression')            # Set the title of the plot.
plt.xlabel('X')                                  # Label the x-axis.
plt.ylabel('y')                                  # Label the y-axis.
plt.legend()                                     # Display the legend on the plot.
plt.show()                                       # Render and display the plot.
```

```python
import numpy as np                                 # Import NumPy for numerical operations.
from sklearn.linear_model import LogisticRegression# Import the Logistic Regression model.

X      = np.array([[0], [1], [2], [3]])            # Feature data: 4 samples, one feature each.
y      = np.array([0, 0, 1, 1])                    # Target data: binary classes (0 or 1) for each sample.
model  = LogisticRegression()                      # Initialize the Logistic Regression model.
model.fit(X, y)                                    # Train the model using the provided data.
predictions = model.predict(X)                     # Predict class labels for the input features.
print("Predictions:", predictions)                 # Print the predicted class labels.
```
